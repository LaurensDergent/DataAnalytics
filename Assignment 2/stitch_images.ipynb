```python
from PIL import Image


def stitch_images(image_location, output_path):
    # Check if the length of the list exceeds 20
    if len(image_location) > 20:
        # If it does, keep only the first 20 elements
        image_location = image_location[:20]

    # Open all images
    images = [Image.open('D:/images/images/' + path) for path in image_location]
    i=0
    # Duplicate images if the length is less than 20
    while len(images) < 20:
        # Choose the index of the image to duplicate (wrap around if necessary)
        index_to_duplicate = i
        # Duplicate the image
        duplicated_image = images[index_to_duplicate].copy()
        # Append the duplicated image to the list
        images.append(duplicated_image)
        i += 1

    # Trim the list to ensure it contains only 20 images
    images = images[:20]
    # Get the dimensions of the first image
    width, height = images[0].size

    # Create a new image with width equal to the sum of widths of all images and height equal to the height of the first image
    stitched_image = Image.new('RGB', (sum(im.width for im in images), height))

    # Paste each image into the stitched image
    x_offset = 0
    for img in images:
        stitched_image.paste(img, (x_offset, 0))
        x_offset += img.width

    # stitched_image.show()
    # Save the stitched image
    stitched_image.save(output_path)

# Example usage
# output_path = 'D:/images/stitched.jpg'
# image_location = image_strings
# stitch_images(image_location, output_path)
```

```python

# Create a list of dictionaries for DataFrame
for index, game in selected_games[0:100].iterrows():
    image_location = game['screenshots']
    new_name = image_location[0]
    output_path = 'D:/images/stitched_3/'+ str(new_name)  
    print(output_path)
    print(image_location)
    stitch_images(image_location, output_path)


#following image did not get stitched: index 12281  2855830_ss_3a3c838324785db1625672817cf7b4dbb59b1b95.1920x1080.jpg(.jpg)
```
